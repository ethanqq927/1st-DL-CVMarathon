### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
import requests,json
url = "https://api.github.com/events"

r = requests.get(url)
r_text = r.text
print(r_text)
j_text = json.loads(r_text)
print(j_text)
```
