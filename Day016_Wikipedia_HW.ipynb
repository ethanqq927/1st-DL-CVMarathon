## 範例：練習是從Wikipedia中爬取文章。先定義一個搜尋的關鍵字，擷取該關鍵字詞的文章。

```python
import requests
import re
from bs4 import BeautifulSoup
import os
```

### 先定義一個我們想搜尋的字詞，並將它轉換成UTF-8編碼後的URL

```python
input_keyword = "斯德哥爾摩症候群"  # 這裡可以自己定義有興趣的關鍵字

utf8_url = repr(input_keyword.encode('UTF-8')).upper()  # 編碼成UTF-8並轉成大寫字元
print(utf8_url)
utf8_url = utf8_url.replace("\\X", "%")
print(utf8_url)# 用 '%' 取代 '\X'
print("%s: %s" % (input_keyword, utf8_url[2:-1:1]))     # 擷取中間的編碼結果

# 組成Wiki關鍵字搜尋的網址格式
root_keyword_link = '/wiki/' + utf8_url[2:-1:1]
print(root_keyword_link)
```

### 範例2：從爬取的文章內容中，擷取出有外部連結的關鍵字。這些關鍵字在文章中是以藍色字體顯示，會連到外部的網頁，並解釋其內容。

```python
for ext_link in content:
    a_tag = ext_link.find_all('a', href=re.compile("^(/wiki/)((?!;)\S)*$"))
    #print(a_tag)
    if len(a_tag) > 0:
        for links in a_tag:
            link = links["href"]       # 外部連結的網址
            keyword = links.text  # 外部連結的中文名稱
            print("外部連結: [{}] {}".format(keyword, link))
```

## 作業：接下來定義一個爬蟲函數，這個函數的主要工作為：
### (1) 爬取當前關鍵字的解釋，並存入檔案(因為文章內容太多會佔滿整個頁面，所以存程檔案，方便後續檢視)
### (2) 萃取出當前關鍵字所引用的外部連結，當作新的查詢關鍵字
### (3) 把第(2)擷取到的關鍵字當作新的關鍵字，回到第(1)步，爬取新的關鍵字解釋。

```python

def WikiArticle(key_word_link, key_word, recursive):
    
    if (recursive <= max_recursive_depth):
        print("遞迴層[%d] - (%s) %s" % (recursive, key_word, key_word_link))

        proxies = {
         "http": "http://127.0.0.1:1081",
         "https": "http://127.0.0.1:1081",
        }

        # 模擬封包的標頭
        headers = {
        'authority': 'zh.wikipedia.org',
        'method': 'GET',
        'path': '/wiki/' + key_word_link,
        'scheme': 'https',
        'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
        'accept-encoding': 'gzip, deflate, br',
        'accept-language': 'en-TW,en-US;q=0.9,en;q=0.8,zh-TW;q=0.7,zh;q=0.6',
        'cookie': 'TBLkisOn=0; GeoIP=TW:CHA:Chang-hua:24.07:120.54:v4; zhwikimwuser-sessionId=0f31bf804b521b6ca19d; WMF-Last-Access=13-Jul-2020; WMF-Last-Access-Global=13-Jul-2020',
        'dnt': '1',
        #'if-modified-since': 'Tue, 04 Jun 2019 12:03:22 GMT',
        'referer': 'https://zh.wikipedia.org/w/index.php?search=%E6%96%AF%E5%BE%B7%E7%A7%91%E7%88%BE%E6%91%A9%E7%97%87%E5%80%99%E7%BE%A4&title=Special:%E6%90%9C%E7%B4%A2&go=%E5%9F%B7%E8%A1%8C&ns0=1&searchToken=7s3q8uo2d1mutwvrmuupoxyny',
        'upgrade-insecure-requests': '1',
        'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36 Edg/83.0.478.56'
        }

        page = requests.get('http://whatsmyuseragent.org/', headers=headers,proxies=proxies)
        soup = BeautifulSoup(page.text, 'html5lib')
        ip = soup.select('.ip-address .intro-text')
        print(ip[0].text)



        url = 'https://zh.wikipedia.org' + key_word_link  # 組合關鍵字查詢URL
        resp = requests.get(url, headers=headers,proxies=proxies)
        resp.encoding = 'utf-8'

        html_soup = BeautifulSoup(resp.text, "lxml")
        content = html_soup.find(id='mw-content-text').find_all('p')

        #
        # Part 1: 請參考範例1，爬取當前關鍵字的文章內容。
        #         因為內容太多，我們把它寫入檔案，並以關鍵字作為檔案名稱，以便稍後查閱內容。
        #         請先建立一個名為"WikiArticle"的資料夾，爬取到的文章內容會放在這個資料夾底下。
        #

        # 決定要儲存的資料夾
        output_dir = f'./WikiArticle/'

        # 假如資料夾不存在就新增一個資料夾
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        with open(f"./WikiArticle/{key_word}.txt", 'w') as f:
            for c in content:
                f.writelines(c.text)
            print("文字內容儲存完畢")


        #
        # Part 2: 請參考範例2，萃取出本篇文章中所延伸引用的外部連結，並儲存在external_link_dict
        #
        external_link_dict = dict({})

        for ext_link in content:
            a_tag = ext_link.find_all('a', href=re.compile("^(/wiki/)((?!;|#)\S)*$"))
            #print(a_tag)
            if len(a_tag) > 0:
                for links in a_tag:
                    link = links["href"]       # 外部連結的網址
                    keyword = links.text  # 外部連結的中文名稱
                    #print("外部連結: [{}] {}".format(keyword, link))
                    external_link_dict[keyword]=link

            #print(external_link_dict)

        #
        # Part 3: 將Part 2所收集的外部連結，當作新的關鍵字，繼續迭代深入爬蟲
        #

        if (len(external_link_dict) > 0):
            recursive = recursive + 1  #遞迴深度加1

            for k, v in external_link_dict.items():
                WikiArticle(v, k, recursive)
```
