### 先搜尋全部的電影代號(ID)資訊

```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd

# 查看目前上映那些電影，並擷取出其ID資訊
url = 'https://movies.yahoo.com.tw/'
resp = requests.get(url)
resp.encoding = 'utf-8'
print(resp.text)
with open("./resp.txt","w") as f:
    f.writelines(resp.text)
```

```python
soup = BeautifulSoup(resp.text, 'lxml')
html = soup.find("select", attrs={'name':'movie_id'})
#print(html)
print(type(html))

movie_item = html.find_all("option",attrs={'data-name':re.compile('.*')})
print(type(movie_item))
print(movie_item)
```

```python
id_list = []
title_list = []
for info in movie_item:
    print(info)
    print(type(info))
    #print("Movie: %s, ID: %s" % (p["data-name"], p["value"]))
    #print("Movie: {}, ID: {}".format(info["data-name"], info["value"]))
    id_list.append(info["value"])
    title_list.append(info["data-name"])

df_movie_id = pd.DataFrame()
df_movie_id["電影"]=title_list
df_movie_id["ID"]=id_list
df_movie_id.to_csv("./movie_id.csv",encoding="utf-8-sig")

df_movie_id
```

### 最後指定觀看的日期，查詢並列印出放映的電影院、放映類型(數位、3D、IMAX 3D...)、放映時間等資訊。

```python
# 選定要觀看的日期
"""注意日期過期（可填當天，明天，後天）"""

date1 = "2020-07-13"
```

```python
# 向網站發送請求，獲取上映的電影院及時間資訊
url = "https://movies.yahoo.com.tw/ajax/pc/get_schedule_by_movie"
print(movie_id)
print(date1)
print(area_id)

payload = {'movie_id':str(movie_id),
           'date':date1,
           'area_id':str(area_id),
           'theater_id':'',
           'datetime':'',
           'movie_type_id':''}

resp = requests.get(url, params=payload)
print(resp.url)

#print(resp.json()['view'])  # 若有需要，列印出json原始碼
#print(resp.json())
json_data = resp.json()
#print(json_data)
print(json_data['view'])
```

```python
soup = BeautifulSoup(json_data['view'],'lxml')
html_elem = soup.find_all("ul", attrs={'data-theater_name':re.compile(".*")})
print(html_elem)
```

```python
'''
  試著從上一步驟回傳的電影院資料中，擷取電影院名稱、影片放映類型以及時間表
  Your code here.
'''

theater_list = []
cate_list= []
schedule_list = []

print(html_elem)
```

```python
#每個the就是一間戲院
for the in html_elem:
    #print(info)
    theater = the.find("li",attrs={"class":"adds"})
    #print(theater)
    print("電影院： {}".format(theater.find("a").text))

    info = the.find_all(class_="gabtn")
    #print(info)
    for i in info:
        #print(i)
        print(i["data-movie_time"],i["data-movie_type"])
        theater_list.append(theater.find("a").text)
        schedule_list.append(i["data-movie_time"])
        cate_list.append(i["data-movie_type"])

    print("====================")

```

```python


df = pd.DataFrame()

df["戲院"] = theater_list
df["類型"] = cate_list
df["時刻"] = schedule_list
df.to_csv("./movie_schedule.csv",encoding="utf-8-sig")
df
```
