```python
import requests
import json

url = "https://www.dcard.tw/service/api/v2/forums?nsfw=true"


headers ={
    'Referer': 'https://www.dcard.tw/f',
'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36 Edg/83.0.478.56'
}

resp = requests.get(url,headers = headers)
#resp.encoding = "utf-8"
for d in json.loads(resp.text):
    print(d["name"])
```
