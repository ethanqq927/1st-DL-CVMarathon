## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import time
import pandas as pd

yesterday_time = datetime.now() - timedelta(1)
yesterday_date = yesterday_time.strftime("%Y/%m/%d")
print("昨天日期：", yesterday_date)

browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
last_height = browser.execute_script("return document.body.scrollHeight")

while True:

    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "lxml")

    new_height = browser.execute_script("return document.body.scrollHeight")

    #已經到頁面底部
    if new_height == last_height:
        break
    last_height = new_height

    if yesterday_date in soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        print("=======已到設定日期=======")
        break
    else:
        print("目前畫面最下面的新聞：", soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text,soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find_all('a')[-1].text)
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "lxml")

news ={"日期時間":[],"標題":[],"連結":[]}

for d in soup.find(class_="part_list_2").find_all('h3'):
    if yesterday_date in d.find(class_="date").text:
        pass
    else:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
        news["日期時間"].append(d.find(class_="date").text)
        news["標題"].append(d.find_all('a')[-1].text)
        news["連結"].append("https://www.ettoday.net" + d.find_all('a')[-1]["href"])

df_news=pd.DataFrame(news)
df_news

```

## 2. 取出現在時間兩小時內的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import time
import pandas as pd

browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

two_hours_ago = datetime.now() - timedelta(hours = 2)
print(two_hours_ago)
two_hours_ago_time = two_hours_ago.strftime("%Y/%m/%d %H:%M")
print("兩小時前時間：", two_hours_ago_time)
```

```python
last_height = browser.execute_script("return document.body.scrollHeight")

go=True

while go:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "lxml")

    new_height = browser.execute_script("return document.body.scrollHeight")

    #已經到頁面底部
    if new_height == last_height:
        print("已經到頁面最底部，程序停止")
        break

    last_height = new_height

    for d in soup.find(class_="part_list_2").find_all('h3'):
        #已經超出兩小時
        if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < two_hours_ago:
            print("已經超出兩個小時，程序停止")
            go = False
            break

        else:
            print("目前畫面最下方文章的日期時間為：",d.find_all(class_="date")[-1].text)

```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "lxml")

news ={"日期時間":[],"標題":[],"連結":[]}

for d in soup.find(class_="part_list_2").find_all('h3'):
    if two_hours_ago_time in d.find(class_="date") :
        pass
    else:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
        news["日期時間"].append(d.find(class_="date").text)
        news["標題"].append(d.find_all('a')[-1].text)
        news["連結"].append("https://www.ettoday.net" + d.find_all('a')[-1]["href"])
import pandas as pd

df_news=pd.DataFrame(news)
df_news
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import time
import pandas as pd
from selenium.webdriver.support.ui import Select


browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
three_days_ago_time = datetime.now() - timedelta(3)
print("取得三天前時間：", three_days_ago_time)

three_days_ago_month = three_days_ago_time.strftime("%m").replace("0","")
three_days_ago_day = three_days_ago_time.strftime("%d").replace("0","")
print("取得三天前的（月）：", three_days_ago_month)
print("取得三天前的（日）：", three_days_ago_day)

three_days_ago_15_clock = datetime.strptime(three_days_ago_time.strftime("%Y/%m/%d") + ' 20', '%Y/%m/%d %H')
three_days_ago_17_clock = datetime.strptime(three_days_ago_time.strftime("%Y/%m/%d") + ' 23', '%Y/%m/%d %H')
print("取得昨天下午三點：", three_days_ago_15_clock)
print("取得昨天下午五點：", three_days_ago_17_clock)
```

```python
# 選擇日期&點選進入頁面
selM = Select(browser.find_element_by_id("selM"))
selM.select_by_value(three_days_ago_month)


selD = Select(browser.find_element_by_id("selD"))
selD.select_by_value(three_days_ago_day)
browser.find_element_by_id('button').click()
```

```python
# 利用 selenium 持續執行下滑的動作，直到取得要求時間內的新聞，表示取得完畢

last_height = browser.execute_script("return document.body.scrollHeight")

if_go=True
while if_go:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "lxml")

    new_height = browser.execute_script("return document.body.scrollHeight")

    #檢查是否已經到頁面底部
    if new_height == last_height:
        print("已經到頁面最底部，程序停止")
        break

    last_height = new_height


    for d in soup.find(class_="part_list_2").find_all('h3'):
        #已經早於指定時間
        if datetime.strptime(d.find_all(class_="date")[-1].text, '%Y/%m/%d %H:%M') < three_days_ago_15_clock:
            print(f"已經早於{three_days_ago_15_clock}，程序停止")
            if_go = False
            break
    print("目前畫面最下方文章的日期時間為：",soup.find(class_="part_list_2").find_all('h3')[-1].find_all(class_="date")[-1].text)



```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "lxml")

news ={"日期時間":[],"標題":[],"連結":[]}

for d in soup.find(class_="part_list_2").find_all('h3'):
    if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < three_days_ago_15_clock  :
        pass

    elif datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') > three_days_ago_17_clock:
        pass

    else:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
        news["日期時間"].append(d.find(class_="date").text)
        news["標題"].append(d.find_all('a')[-1].text)
        news["連結"].append("https://www.ettoday.net" + d.find_all('a')[-1]["href"])

df_news=pd.DataFrame(news)
df_news

```
