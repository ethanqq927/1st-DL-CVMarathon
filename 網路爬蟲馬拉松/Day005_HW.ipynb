```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
import requests, json
r_pop = requests.get('https://www.dcard.tw/_api/forums/job/posts?popular=true')
response_pop = r_pop.text
data_pop = json.loads(response_pop)

sum_pop_commentcount=0
sum_pop_likecount=0
for d in data_pop:
    sum_pop_commentcount +=d['commentCount']
    sum_pop_likecount += d['likeCount']
avg_pop_commentcount = sum_pop_commentcount/len(data_pop)
avg_pop_likecount = sum_pop_likecount/len(data_pop)
print(f"ç†±é–€æ–‡ç« å¹³å‡ç•™è¨€æ•¸ï¼š{avg_pop_commentcount}")
print(f"ç†±é–€æ–‡ç« å¹³å‡æš—æŒ‰è®šæ•¸ï¼š{avg_pop_likecount}")

r_new = requests.get('https://www.dcard.tw/_api/forums/job/posts?popular=false')
response_new = r_new.text
data_new = json.loads(response_new)

sum_new_commentcount=0
sum_new_likecount=0
for d in data_new:
    sum_new_commentcount +=d['commentCount']
    sum_new_likecount += d['likeCount']
avg_new_commentcount = sum_new_commentcount/len(data_new)
avg_new_likecount = sum_new_likecount/len(data_new)
print(f"æœ€æ–°æ–‡ç« å¹³å‡ç•™è¨€æ•¸ï¼š{avg_new_commentcount}")
print(f"æœ€æ–°æ–‡ç« å¹³å‡ç•™è¨€æ•¸ï¼š{avg_new_likecount}")



```
