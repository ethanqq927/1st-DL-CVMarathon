```python
soup = BeautifulSoup(resp.text,"html5lib")

post_css=".part_list_2"
cate_css = ".tag c_news"

elem = soup.select(post_css)
title_list = []
date_list = []
cate_list = []
for e in elem:
    title_list = [title.text for title in e.select("a")]
    #print(title_list)
    date_list = [date.text for date in e.select(".date")]
    #print(date_list)
    cate_list = [cate.text for cate in e.select("em")]
    #print(cate_list)

df = pd.DataFrame()
df["title"]=title_list
df["category"]= cate_list
df["date"]=date_list

df
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
df.sort_values("category")
```

### ③ 哪一個類別的文章最多

```python
name_highest = df["category"].value_counts().idxmax()
count_highest = df["category"].value_counts().max()
print(f"文章出現最多次的類別是：{name_highest}類:{count_highest}次")
```
